### **Get video ID from URL**

```
ids = pd.DataFrame(history)
```

### **Drop the columns which are not required**

```
ids = ids.drop(['header','subtitles','products','details','description'],axis=1)
```

```
ids.head()
```

### **Helper functions to get id and title**

```
def clean(x):
    x = str(x).replace("https://www.youtube.com/watch?v=","")
    return str(x)
ids['video_id'] = ids['titleUrl'].apply(clean)
```

```
def clean1(x):
    x = str(x).replace("Watched ","")
    return str(x)
ids['title'] = ids['title'].apply(clean1)
```

```
ids.head()
```

## **Get the category id, channel name and the video title**

```
ids['category_id'] = ''
ids['channel'] = ''
```

```
def get_data(vid):
  try:
    request = youtube.videos().list(part="snippet",id=vid)
    response = request.execute()
    snippet = response['items'][0]['snippet']
    res = []
    res.append(snippet['categoryId'])
    res.append(snippet['channelTitle'])
    return res
  except Exception as e:
            traceback.print_exc()

```

```
count=5000 
for index, row in ids.iloc[5000:len(ids)].iterrows():
    try:
            print(count)
            count+=1
            vid = row['video_id']
            res=get_data(vid)
            ids.at[index,'category_id']=res[0]
            ids.at[index,'channel']=res[1]
    except Exception as e:
        print(count)
        traceback.print_exc()
        ids.to_csv('data_5k_15k.csv')
        print('last save at--> ')
        print(count)
        continue
print(count)            
ids[5000:len(ids)].to_csv('complete_data_5000_15k.csv')
```
